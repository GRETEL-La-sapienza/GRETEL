```python

import os
# THIS FUNCTION BUILDS NEW INPUT DATASET FOR A REDUCED GRAPH (with less than 50 nodes dimension)
#  print current working directory
print("Directory di lavoro corrente:", os.getcwd())

# change of working directory on your notebook
os.chdir('/Users/bruzzese/GRETEL/data/datasets/github_stargazers/')
print("Directory modificata in:", os.getcwd())

def analyze_graphs(filename):
    graph_count = {}
    with open(filename, 'r') as file:
        for line in file:
            graph_id = int(line.strip())
            if graph_id in graph_count:
                graph_count[graph_id] += 1
            else:
                graph_count[graph_id] = 1
    
    return graph_count

# File path for the dataset
filename = 'github_stargazers_graph_indicator.txt'

# Execute the function and print the results
results = analyze_graphs(filename)
for graph_id, node_count in results.items():
    print(f"Graph ID: {graph_id}, Number of Nodes: {node_count}")


def analyze_and_extract_graphs(filename_indicator, filename_A, filename_labels, output_dir, max_nodes=50):
    # Analyze the original dataset in order to count the nodes 
    graph_counts = analyze_graphs(filename_indicator)
    
    # identify the graphs with al most max_nodes nodes
    small_graphs = {graph_id for graph_id, count in graph_counts.items() if count <= max_nodes}

    # Prepare paths for output files 
    output_file_nodes = os.path.join(output_dir, 'github_stargazers_graph_indicators.txt')
    output_file_edges = os.path.join(output_dir, 'github_stargazers_A.txt')
    output_file_labels = os.path.join(output_dir,'github_stargazers_graph_labels.txt')

    # Extrapolate the nodees for selected grapsh 
    with open(filename_indicator, 'r') as file, open(output_file_nodes, 'w') as out_file:
        for line in file:
            graph_id = int(line.strip())
            if graph_id in small_graphs:
                out_file.write(line)

    # Extrapolate couple of nodes  (edges) for selected grapsh 
    with open(filename_A, 'r') as file, open(output_file_edges, 'w') as out_file:
        for line in file:
            node1, node2 = map(int, line.split(','))
            if node1 in small_graphs and node2 in small_graphs:
                out_file.write(line)


    # Extrapolate the labels for selected graphs
    with open(filename_labels, 'r') as file, open(output_file_labels, 'w') as out_file:
        for graph_id, line in enumerate(file, start=1):  # start counting from 1
            label = line.strip()  # remove empty spaces or characters of new line
            if graph_id in small_graphs:
                out_file.write(label + '\n')
    

# Paths of output files and folder
filename_indicator = 'github_stargazers_graph_indicator.txt'
filename_A = 'github_stargazers_A.txt'
filename_labels = 'github_stargazers_graph_labels.txt'
output_dir = 'small_graphs'



#  Execute the function
analyze_and_extract_graphs(filename_indicator, filename_A, filename_labels, output_dir)


```
